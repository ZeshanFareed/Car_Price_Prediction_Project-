# Converting the Object data into the numrical Through Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['Car_Name'].unique()
```

```python
Car_Name_lb = LabelEncoder()

df['Car_Name'] = Car_Name_lb.fit_transform(df['Car_Name'])
```

```python
df['Fuel_Type'].unique()
```

```python
Fuel_Type_lb = LabelEncoder()

df['Fuel_Type'] = Fuel_Type_lb.fit_transform(df['Fuel_Type'])
```

# Encoding the Selling_type Object data

```python
df['Selling_type'].unique()
```

```python
Selling_type_lb = LabelEncoder()

df['Selling_type'] = Selling_type_lb.fit_transform(df['Selling_type'])
```

# Encoding the Transmission  Object data

```python
df['Transmission'].unique()
```

```python
Transmission_lb = LabelEncoder()

df['Transmission'] = Transmission_lb.fit_transform(df['Transmission'])
```

# Feature and Target Variable Selection

```python
input_data_x = df.iloc[: , : -1]

output_data_y = df['Selling_Price']
```

# Scaling the input_data_x

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()

input_data_x = pd.DataFrame(ss.fit_transform(input_data_x) , columns=input_data_x.columns)
```

```python
input_data_x
```

# Train-Test Split for Model Evaluation

```python
from sklearn.model_selection import train_test_split 

X_train , X_test , y_train , y_test = train_test_split(input_data_x , output_data_y , test_size=0.2 , random_state=42)
```

# All possible Supervised Machine Algorithms are apply 

```python
from sklearn.linear_model import LinearRegression , Ridge , Lasso , ElasticNet 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV

from sklearn.metrics import mean_squared_error , mean_absolute_error
```

# Training and Evaluating Random Forest Regressor

```python
RF = RandomForestRegressor()

RF.fit(X_train , y_train)  
```

## Model Accuracy: Training and Testing Scores

```python
train_score = RF.score(X_train, y_train) * 100
test_score = RF.score(X_test, y_test) * 100
```

```python
print(f"Train Score: {train_score:.2f}%")
print(f"Test Score: {test_score:.2f}%")
```

## Plot training and testing accuracy

```python
plt.figure(figsize=(8, 5))
bars = ['Training Accuracy', 'Testing Accuracy']
values = [train_score, test_score]

plt.bar(bars, values, color=['blue', 'orange'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy (%)')
plt.ylim([0, 100])
plt.show()
```

## Model Error Metrics: MSE and MAE

```python
mse = mean_squared_error(y_test, RF.predict(X_test))
mae = mean_absolute_error(y_test, RF.predict(X_test))
```

```python
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

## Plot MSE and MAE

```python
plt.figure(figsize=(8, 5))
error_metrics = ['Mean Squared Error', 'Mean Absolute Error']
values = [mse, mae]

plt.bar(error_metrics, values, color=['green', 'red'])
plt.title('Error Metrics')
plt.ylabel('Error Value')
plt.show()
```
